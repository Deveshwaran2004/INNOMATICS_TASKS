### Frequency and Probability distribution of numerical columns in the dataset

#### Frequency Distribution
A frequency distribution is a summary of how often each value occurs in a dataset. It shows the number of times each possible value of a variable appears. For example, if you have a list of exam scores, a frequency distribution would tell you how many students received each possible score.

#### Probability Distribution
A probability distribution describes how the values of a random variable are distributed. It provides the probabilities of different possible outcomes. For example, in a dice roll, the probability distribution would show that each number (1 through 6) has an equal chance of occurring, which is (\frac{1}{6}).

#### Why Perform These on Numerical Columns?
Performing frequency and probability distributions on numerical columns helps in:

##### Understanding Data Patterns: 
It reveals the distribution and spread of data, helping to identify common values, outliers, and overall trends.
##### Statistical Analysis: 
It forms the basis for many statistical analyses, such as calculating mean, median, mode, variance, and standard deviation.
##### Probability Calculations:
It helps in determining the likelihood of different outcomes, which is crucial for predictive modeling and risk assessment.

```python
# Function to create side-by-side histograms for specific numerical columns
def plot_side_by_side_histograms(df1, df2, title1, title2, columns):
    # Plotting for specified columns
    for column in columns:
        if column in df1.columns and column in df2.columns:
            fig, axes = plt.subplots(1, 2, figsize=(18, 6))  # 1 row, 2 columns
            
            # Plot histogram for df1 (df_amcat_left)
            sns.histplot(df1[column].dropna(), kde=True, ax=axes[0])  # Drop NA values to avoid errors
            axes[0].set_title(f'Probability and Frequency Distribution of {column} ({title1})')
            axes[0].set_xlabel(column)
            axes[0].set_ylabel('Frequency')
            
            # Plot histogram for df2 (df_amcat_present)
            sns.histplot(df2[column].dropna(), kde=True, ax=axes[1])  # Drop NA values to avoid errors
            axes[1].set_title(f'Probability and Frequency Distribution of {column} ({title2})')
            axes[1].set_xlabel(column)
            axes[1].set_ylabel('Frequency')
            
            # Adjust layout and show
            plt.tight_layout()
            plt.show()

# Call the function to plot side-by-side histograms for specified columns
columns_to_plot = ['12percentage', '10percentage', 'collegeGPA', 'Salary']
plot_side_by_side_histograms(df_amcat_left, df_amcat_present, 'Left', 'Present', columns_to_plot)
```

```python
def plot_degree_and_specialization(df1, df2, title1, title2):
    # Plot for Degree
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))  # 1 row, 2 columns for Degree
    
    # Plot for Degree in df1
    sns.countplot(y=df1['Degree'], order=df1['Degree'].value_counts().index, ax=axes[0])
    axes[0].set_title(f'Degree Distribution ({title1})')
    axes[0].set_xlabel('Frequency')
    axes[0].set_ylabel('Degree')
    
    # Plot for Degree in df2
    sns.countplot(y=df2['Degree'], order=df2['Degree'].value_counts().index, ax=axes[1])
    axes[1].set_title(f'Degree Distribution ({title2})')
    axes[1].set_xlabel('Frequency')
    axes[1].set_ylabel('')  # No ylabel for this subplot to keep things clean
    
    # Adjust layout for Degree
    plt.tight_layout()
    plt.show()
    
    # Find the highest degree in each dataset
    highest_degree_df1 = df1['Degree'].value_counts().idxmax()
    highest_degree_df2 = df2['Degree'].value_counts().idxmax()
    
    # Filter the datasets for rows with the highest degree
    df1_highest_degree = df1[df1['Degree'] == highest_degree_df1]
    df2_highest_degree = df2[df2['Degree'] == highest_degree_df2]
    
    # Plot for Specialization (for the highest degree)
    fig, axes = plt.subplots(1, 2, figsize=(18, 6))  # 1 row, 2 columns for Specialization
    
    # Plot Specialization for the highest degree in df1
    sns.countplot(y=df1_highest_degree['Specialization'], 
                  order=df1_highest_degree['Specialization'].value_counts().index, 
                  ax=axes[0])
    axes[0].set_title(f'Specialization Distribution for {highest_degree_df1} ({title1})')
    axes[0].set_xlabel('Frequency')
    axes[0].set_ylabel('Specialization')
    
    # Plot Specialization for the highest degree in df2
    sns.countplot(y=df2_highest_degree['Specialization'], 
                  order=df2_highest_degree['Specialization'].value_counts().index, 
                  ax=axes[1])
    axes[1].set_title(f'Specialization Distribution for {highest_degree_df2} ({title2})')
    axes[1].set_xlabel('Frequency')
    axes[1].set_ylabel('')  # No ylabel for this subplot to keep things clean
    
    # Adjust layout for Specialization
    plt.tight_layout()
    plt.show()

# Call the function to plot Degree and Specialization side-by-side
plot_degree_and_specialization(df_amcat_left, df_amcat_present, 'Left', 'Present')
```

```python
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(df_amcat['Gender'], df_amcat['Specialization'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-Square Test Results:")
print(f"Chi2 Value: {chi2}")
print(f"P-Value: {p}")
print("Contingency Table:")
print(contingency_table)

# Interpretation of the results
alpha = 0.05  # significance level
if p < alpha:
    print("Reject the null hypothesis: There is a significant relationship between Gender and Specialization.")
else:
    print("Fail to reject the null hypothesis: There is no significant relationship between Gender and Specialization.")
```
